```python
import pandas as pd
import numpy as np
```

```python
# load the data
path="/content/modecraftdata.csv"
retail=pd.read_csv(path)
retail.head(5)
```

```python
retail.shape
```

```python
# check for missing values or empty strings in dataset
((retail.isnull()) | (retail == '')).values.any()
```

```python
# replace the empty strings with nan so you can drop them
retail.replace('', np.nan, inplace=True)
```

```python
# getting rid of the blanks/null values
retail.dropna(inplace=True)
```

```python
retail.shape
```

```python
# some descriptions have ?, we're getting rid of that
retail = retail[~retail['Description'].str.contains('\?', regex=True)]
```

```python
retail.shape
```

```python
# we want to get rid of negative values so our revenue calculations aren't messed up
retail = retail[(retail['UnitPrice'] > 0) & (retail['Quantity'] > 0)]
```

```python
retail.shape
```

```python
# now add more columns -- revenue column
retail["Revenue"] = retail["Quantity"]*retail["UnitPrice"]
retail.head(5)
```

```python
# Convert InvoiceDate to datetime
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
```

```python
retail.head(5)
```

```python
# day of week
retail['DayOfWeek'] = retail['InvoiceDate'].dt.day_name()
```

```python
# function to get the time of day (morning, afternoon, night)
def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

retail['TimeOfDay'] = retail['InvoiceDate'].dt.hour.apply(get_time_of_day)
```

```python
retail['WeekdayOrWeekend'] = retail['InvoiceDate'].dt.weekday.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')
```

```python
retail['Quarter'] = retail['InvoiceDate'].dt.quarter.apply(lambda x: f"Q{x}")
```

```python
retail['ProductName'] = retail['Description'].str.title()  # Title-case the product name
```

```python
retail.shape
```

```python
retail.head(5)
```

```python
# now uploading the cleaned dataset
retail.to_csv('modekraft_updated.csv', index=False)
```
